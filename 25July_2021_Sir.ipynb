```python
data = {"Country":["US","US", "US","IND","IND","UK","UK"],
       "Cases":[100,200,300,40,50,600,345]}
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
df.groupby("Country")
```

```python
country_group=df.groupby("Country")
```

```python
country_group.mean()
```

```python
df.groupby("Country").mean()
```

```python
df.groupby("Country").sum()
```

```python
df.groupby("Country").first()
```

```python
#The groupby command ca be used for filtering as well
df.groupby("Country").get_group("IND")
```
