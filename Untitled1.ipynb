# Grouping of the data

```python
data = {"Country":["US","US", "US","IND","IND","UK","UK"],
       "Cases":[100,200,300,40,50,600,345]}
```

```python
df = pd.DataFrame(data);
df
```

```python
country_grp = df.groupby("Country")
```

```python
country_grp.mean()
```

```python
country_grp.median()
```

```python
country_grp.sum()
```

```python
country_grp.first()
```

```python
# The grouping can be done for filtering as well
df.groupby("Country").get_group("IND")
```

```python
df.groupby(["Country","Gender"]).mean()
```

```python
df.groupby("Country").describe()
```

# Merging of 2 DataFrames using Joins

```python
df1= pd.DataFrame({"ID":[1,2,3,4,5],
                  "Age":[12,13,14,15,17]})


df2 = pd.DataFrame({"ID":[1,2,3,4,5],"Sub":["A","B","C","D","E"]})
df1,df2
```

```python
pd.merge(df1,df2,how="inner",on="ID")
```

```python
pd.merge(df1,df2,how="left",on="ID")
```

```python
pd.merge(df1,df2,how="right")
```

```python
pd.merge(df1,df2,how="cross")
```
